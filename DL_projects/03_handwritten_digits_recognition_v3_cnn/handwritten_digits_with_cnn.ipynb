```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Handwritten digits recognition model using CNN

```python
( X_train, y_train ), ( X_test, y_test ) = tf.keras.datasets.mnist.load_data( )
```

```python
print( X_train.shape )
print( X_test.shape )
```

```python
def plot_sample( X, y, how_many = 16 ):
    if how_many % 2 == 1:
        how_many -= 1
        
    num_rows = int( np.round( np.sqrt( how_many ) ) )
    num_cols = how_many // num_rows
    
    images = X[:how_many]
    labels = y[:how_many]
    
    fig, axes = plt.subplots( num_rows, num_cols, figsize = ( 1.5 * num_cols, 2 * num_rows ) )
    for i in range( how_many ):
        ax = axes[i // num_cols, i % num_cols]
        ax.imshow( images[i], cmap='gray' )
        ax.set_title( "Label: {}".format( labels[i] ) )
    plt.tight_layout( )
    plt.show( )
```

```python
plot_sample( X_train, y_train )
```

```python
# Normalizing data
X_train = X_train.astype( "float" )
X_test = X_test.astype( "float" )

X_train /= 255
X_test /= 255
```

```python
model = tf.keras.Sequential( [
    tf.keras.layers.Conv2D( filters = 8, kernel_size = ( 5, 5 ), activation = "relu",
                            padding = "same", strides = ( 1, 1 ), input_shape = ( 28, 28, 1 ) ),
    tf.keras.layers.MaxPool2D( pool_size = ( 2, 2 ) ),
    tf.keras.layers.Dropout( 0.3 ),
    
    tf.keras.layers.Conv2D( filters = 16, kernel_size = ( 3, 3 ), activation = "relu",
                            padding = "same", strides = ( 1, 1 ) ),
    tf.keras.layers.MaxPool2D( pool_size = ( 2, 2 ) ),
    tf.keras.layers.Dropout( 0.3 ),
    
    tf.keras.layers.Flatten( ),
    tf.keras.layers.Dense( units = 64, activation = "relu" ),
    tf.keras.layers.Dropout( 0.3 ),
    tf.keras.layers.Dense( units = 32, activation = "relu" ),
    tf.keras.layers.Dropout( 0.3 ),
    tf.keras.layers.Dense( units = 10 )
] )

model.compile( loss = tf.keras.losses.SparseCategoricalCrossentropy( from_logits = True ),
               optimizer = "Adam",
               metrics = ["accuracy"] )
```

```python
model.summary( )
```

```python
history = model.fit( X_train, y_train, epochs = 10, validation_split = 0.15 )
```

```python
model.evaluate( X_test, y_test )
```

```python
plt.figure( figsize = ( 12, 8 ) )
plt.plot( history.history["accuracy"] )
plt.plot( history.history["val_accuracy"] )
plt.title( "model accuracy" )
plt.xlabel( "epochs" )
plt.ylabel( "accuracy" )
plt.legend( ["train", "val"], loc = "upper left" )
plt.show( )
```

```python
plt.figure( figsize = ( 12, 8 ) )
plt.plot( history.history["loss"] )
plt.plot( history.history["val_loss"] )
plt.title( "model loss" )
plt.xlabel( "epochs" )
plt.ylabel( "loss" )
plt.legend( ["train", "val"], loc = "upper left" )
plt.show( )
```

```python
logits = model( X_test )
y_pred = np.array( [np.argmax( i ) for i in tf.nn.softmax( logits )] )
```

```python
def plot_sample_with_preds( X, y, predictions, how_many = 16 ):
    if how_many % 2 == 1:
        how_many -= 1
        
    num_rows = int( np.round( np.sqrt( how_many ) ) )
    num_cols = how_many // num_rows
    
    images = X[:how_many]
    labels = y[:how_many]
    pred = predictions[:how_many]
    
    fig, axes = plt.subplots( num_rows, num_cols, figsize = ( 1.5 * num_cols, 2 * num_rows ) )
    for i in range( how_many ):
        ax = axes[i // num_cols, i % num_cols]
        ax.imshow( images[i], cmap='gray' )
        ax.set_title( "Label: {}, Pred: {}".format( labels[i], pred[i] ) )
    plt.tight_layout( )
    plt.show( )
```

```python
plot_sample_with_preds( X_test, y_test, y_pred, 20 )
```

```python
import seaborn as sns

cfm = tf.math.confusion_matrix( y_test, y_pred, num_classes = 10 )

plt.figure( figsize = ( 12, 8 ) )
sns.heatmap( cfm, annot = True, fmt = 'd' )
plt.xlabel( "Predicted value" )
plt.ylabel( "True value" )
plt.show( )
```
